### Explanation for Cell 1This cell executes: `import os`, `import cv2`, and `import matplotlib.pyplot as plt`.This explanation cell was added to describe what was done in the next cell.

```python
import os
import cv2
import matplotlib.pyplot as plt
import random
```

### Explanation for Cell 2This cell executes: `data_path = "../dataset/imgs/train"`, `classes = os.listdir(data_path)`, and `print("Classes:", classes)`.This explanation cell was added to describe what was done in the next cell.

```python
data_path = "../dataset/imgs/train"
classes = os.listdir(data_path)

print("Classes:", classes)
```

### Explanation for Cell 3This cell executes: `selected_class = random.choice(classes)`, `img_name = random.choice(os.listdir(os.path.join(data_path, selected_class)))`, and `img_path = os.path.join(data_path, selected_class, img_name)`.This explanation cell was added to describe what was done in the next cell.

```python
# pick random class
selected_class = random.choice(classes)

# pick random image
img_name = random.choice(os.listdir(os.path.join(data_path, selected_class)))
img_path = os.path.join(data_path, selected_class, img_name)

# read and convert image
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# show image
plt.imshow(img)
plt.title(f"Class: {selected_class}")
plt.axis("off")
```
